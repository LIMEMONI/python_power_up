```python
import numpy as np
```

```python
a = np.arange(6).reshape(2,3)
b = np.arange(3)
print(a)
print(b)
print(a*b)
```

```python
a = np.arange(6).reshape(2,3)
b = np.arange(2).reshape(-1,2)
print(a)
print(b)
print(a*b)
```

```python
a = np.arange(56).reshape(8,1,7,1)
b = np.arange(15).reshape(3,1,5)
# print(a)
# print(b)
print(a*b)
```

NumPy에서 요소별 곱셈(element-wise multiplication)을 수행하기 위해서는 두 배열의 형태가 브로드캐스팅(broadcasting) 규칙을 만족해야 합니다.

브로드캐스팅 규칙에 따라, 두 배열의 각 차원은 서로 호환되어야 합니다. 
호환된다는 것은 해당 차원이 같거나 둘 중 하나가 1이어야 한다는 의미입니다. 

차원이 부족한 경우 오른쪽으로 정렬 1을 차원으로 넣어버림
